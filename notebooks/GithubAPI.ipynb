CHiamata API 

```python
# Installazione: pip install PyGithub
from github import Github
import json

# 1. Configurazione
TOKEN = ""
REPO_NAME = "crewAIInc/crewAI"

g = Github(TOKEN)
repo = g.get_repo(REPO_NAME)

# 2. Recupera le ultime 10 esecuzioni (o quante ne preferisci)
runs = repo.get_workflow_runs()[:10]

# Lista che conterrà i dati grezzi completi
all_runs_raw_data = []

for run in runs:
    # .raw_data contiene l'intero dizionario originale restituito dall'API di GitHub
    # È esattamente quello che vedresti facendo una chiamata curl
    all_runs_raw_data.append(run.raw_data)
    
    # Stampa di controllo rapida
    print(f"ID Run: {run.id} | Commit: {run.head_sha} | Salvataggio dati completi...")

# 3. Salvataggio su file JSON
nome_file = "workflow_runs_full_data.json"
with open(nome_file, "w", encoding="utf-8") as f:
    # Salviamo l'intera lista di dizionari grezzi
    json.dump(all_runs_raw_data, f, indent=4, ensure_ascii=False)

print(f"\n✅ Fatto! Il file '{nome_file}' contiene ora tutti i dettagli di ogni run.")
```
